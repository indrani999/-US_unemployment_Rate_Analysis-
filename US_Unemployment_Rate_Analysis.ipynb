###### U.S. Unemployment Rates: A Comprehensive Analysis The U.S. job market, with its dynamic trends and fluctuating unemployment rates, serves as an important barometer for the nation's economic health. By analyzing the unemployment rates based on gender and age groups, we can uncover underlying patterns and address potential disparities.

1. Overall Unemployment Trend The overall unemployment rate in the U.S. has experienced several notable fluctuations, with sharp spikes corresponding to economic recessions. The cyclical nature of unemployment is evident, with rates rising during downturns and falling during growth periods. The significant surge around 2020 is attributed to the economic impacts of the COVID-19 pandemic.

2. Gender-Based Analysis Both men's and women's unemployment rates generally align, indicating that macroeconomic events often impact both genders similarly. However, there are periods of divergence, such as during the 2008 recession, where men were more adversely affected. Around 2020, women experienced a brief but sharper spike, hinting at the pandemic's potential gender-specific impacts.

3. Age Group Analysis - Younger Population Younger age groups (16-19 years) consistently experience higher and more volatile unemployment rates. The age group 16-17 faces the highest unemployment challenges, suggesting a need for more focused interventions for younger job seekers.

4. Age Group Analysis - Older Population For older age groups (25 and above), unemployment rates are more stable and closely packed. The 55+ age group, however, generally exhibits a slightly higher unemployment rate, especially in recent years. This calls for attention to potential challenges faced by older workers in the job market, such as technological adaptation, industry shifts, or age-related biases.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_sex = pd.read_csv('/content/df_sex_unemployment_rates.csv')
df_age = pd.read_csv('/content/df_unemployment_rates.csv')
```

```python
df_sex.head()
```

```python
df_age.head()
```

```python
# Convert 'date' column to datetime format for both dataframes
df_sex['date'] = pd.to_datetime(df_sex['date'])
df_age['date'] = pd.to_datetime(df_age['date'])

# Plotting the overall unemployment rate over time
plt.figure(figsize=(15, 8))
plt.plot(df_sex['date'], df_sex['overall_rate'], label='Overall Unemployment Rate', color='red', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Overall U.S. Unemployment Rate Over Time')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting the unemployment rates for men and women over time
plt.figure(figsize=(15, 8))
plt.plot(df_sex['date'], df_sex['men_rate'], label='Men Unemployment Rate', color='red', linewidth=2)
plt.plot(df_sex['date'], df_sex['women_rate'], label='Women Unemployment Rate', color='black', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rates for Men vs. Women Over Time')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting the unemployment rates for younger age groups over time
plt.figure(figsize=(15, 8))
plt.plot(df_age['date'], df_age['age_16_17_rate'], label='Age 16-17 Unemployment Rate', color='blue', linewidth=2)
plt.plot(df_age['date'], df_age['age_16_19_rate'], label='Age 16-19 Unemployment Rate', color='green', linewidth=2)
plt.plot(df_age['date'], df_age['age_18_19_rate'], label='Age 18-19 Unemployment Rate', color='red', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rates for Younger Age Groups Over Time')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting the unemployment rates for older age groups over time
plt.figure(figsize=(15, 8))
plt.plot(df_age['date'], df_age['age_25_34_rate'], label='Age 25-34 Unemployment Rate', color='blue', linewidth=2)
plt.plot(df_age['date'], df_age['age_35_44_rate'], label='Age 35-44 Unemployment Rate', color='green', linewidth=2)
plt.plot(df_age['date'], df_age['age_45_54_rate'], label='Age 45-54 Unemployment Rate', color='red', linewidth=2)
plt.plot(df_age['date'], df_age['age_55plus_rate'], label='Age 55+ Unemployment Rate', color='purple', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rates for Older Age Groups Over Time')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Extract year from the date for grouping
df_sex['year'] = df_sex['date'].dt.year

# Filter out columns we want to visualize
columns_to_visualize = ['men_rate', 'women_rate', 'men_16_17_rate', 'women_16_17_rate',
                        'men_25_34_rate', 'women_25_34_rate', 'men_55plus_rate', 'women_55plus_rate']

# Group by year and get the average for each category
heatmap_data = df_sex.groupby('year')[columns_to_visualize].mean()

# Plotting the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(heatmap_data.T, cmap="YlGnBu", linewidths=.5)
plt.title('Unemployment Rates Heatmap: Gender & Age Groups Over Years')
plt.xlabel('Year')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

#### This heatmap provides an overview of unemployment rates for various gender and age groups over the years. The color gradient is crucial: Darker shades signify higher unemployment rates, while lighter shades indicate lower rates.

1. Gender Differences: A quick look reveals periods when both men and women experienced varying unemployment rates. This is evident in the shading differences between the men_rate and women_rate rows.

2. Youth Unemployment: Rows representing younger age groups (e.g., men_16_17_rate, women_16_17_rate) consistently display darker shades, emphasizing the higher unemployment challenges faced by these age groups.

3. Age Patterns: Moving to older age categories, the shades generally lighten, suggesting lower unemployment rates for these groups.

4. Historical Perspective: Certain columns, such as 2008 and 2020, are notably darker, reflecting economic challenges during those times.

```python
# Calculating the correlation matrix for selected columns
correlation_matrix = df_sex[columns_to_visualize].corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5, vmin=-1, vmax=1)
plt.title('Correlation Matrix of Unemployment Rates')
plt.tight_layout()
plt.show()
```

```python
# Plotting the spread between men's and women's unemployment rates over time using an area chart
plt.figure(figsize=(15, 8))
plt.fill_between(df_sex['date'], df_sex['men_rate'], df_sex['women_rate'],
                 where=(df_sex['men_rate'] > df_sex['women_rate']),
                 color='blue', alpha=0.5, label='Men > Women')
plt.fill_between(df_sex['date'], df_sex['men_rate'], df_sex['women_rate'],
                 where=(df_sex['men_rate'] <= df_sex['women_rate']),
                 color='orange', alpha=0.5, label='Women >= Men')
plt.plot(df_sex['date'], df_sex['men_rate'], label='Men Unemployment Rate', color='blue', linewidth=2)
plt.plot(df_sex['date'], df_sex['women_rate'], label='Women Unemployment Rate', color='orange', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Spread Between Men\'s and Women\'s Unemployment Rates Over Time')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

##### The area chart provides a visualization of the spread between men's and women's unemployment rates over time:

1. Blue Shading: Indicates periods when the unemployment rate for men was higher than for women. The depth of the shading shows the magnitude of the difference.

2. Orange Shading: Represents periods when the unemployment rate for women was equal to or higher than for men.

3. Recessions & Disparities: Noticeable disparities (either direction) often coincide with economic downturns. For instance, around the 2008 recession, the unemployment rate for men surged more than for women, resulting in a larger blue-shaded area.

4. Recent Trends: Post-2020, we observe that the unemployment rate for women briefly surpassed that for men, leading to an orange-shaded region. This may reflect the gender-specific impacts of the COVID-19 pandemic on employment.

```python
# Calculating the 12-month moving average for overall, men, and women unemployment rates
df_sex['overall_moving_avg'] = df_sex['overall_rate'].rolling(window=12).mean()
df_sex['men_moving_avg'] = df_sex['men_rate'].rolling(window=12).mean()
df_sex['women_moving_avg'] = df_sex['women_rate'].rolling(window=12).mean()

# Plotting the moving averages
plt.figure(figsize=(15, 8))
plt.plot(df_sex['date'], df_sex['overall_moving_avg'], label='Overall Unemployment Rate (12-Month MA)', color='black', linewidth=2)
plt.plot(df_sex['date'], df_sex['men_moving_avg'], label='Men Unemployment Rate (12-Month MA)', color='blue', linewidth=2, linestyle='--')
plt.plot(df_sex['date'], df_sex['women_moving_avg'], label='Women Unemployment Rate (12-Month MA)', color='orange', linewidth=2, linestyle='--')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('12-Month Moving Average of Unemployment Rates Over Time')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

##### The 12-month moving average visualization offers the following insights:

1. Smoothed Trends: The moving average helps in highlighting the longer-term trends in unemployment rates by smoothing out short-term fluctuations and seasonality.

2. Consistent Movements: The moving averages for men and women generally move in tandem, reinforcing that macroeconomic factors impact both genders similarly. However, the subtle differences between the two lines can offer insights into gender-specific challenges or advantages during certain periods.

3. Identifying Prolonged Peaks: While the raw data might show sharp spikes, the moving average can help identify periods where high unemployment persisted for a longer duration, such as the period post-2008 and post-2020

```python
# Preparing data for box plots
boxplot_data = df_sex.melt(id_vars='date', value_vars=columns_to_visualize, var_name='Category', value_name='Unemployment Rate')

# Plotting box plots for different categories
plt.figure(figsize=(15, 10))
sns.boxplot(data=boxplot_data, x='Category', y='Unemployment Rate', palette='viridis')
plt.xticks(rotation=45)
plt.title('Distribution of Unemployment Rates Across Different Categories')
plt.tight_layout()
plt.show()
```

```python
# List of columns for age groups from the df_age dataframe for the stacked area chart
age_columns = ['age_16_17_rate', 'age_18_19_rate', 'age_20_24_rate', 'age_25_34_rate',
               'age_35_44_rate', 'age_45_54_rate', 'age_55plus_rate']

# Plotting the stacked area chart
plt.figure(figsize=(15, 8))
plt.stackplot(df_age['date'], *[df_age[col] for col in age_columns], labels=age_columns, alpha=0.6)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Composition of Unemployment Rates Across Age Groups Over Time')
plt.legend(loc='upper left')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

#### The stacked area chart offers a comprehensive view of how different age groups contribute to the overall unemployment rate:

1. Dominant Age Groups: Age groups such as age_20_24_rate and age_25_34_rate consistently contribute a significant portion of the overall unemployment rate, indicating that these age groups face persistent challenges in the job market.

2. Youth Unemployment: The bottom layers, representing younger age groups (e.g., age_16_17_rate and age_18_19_rate), showcase that while these groups have a higher unemployment rate percentage-wise, their contribution to the total unemployment figure is comparatively smaller. This is because the size of these age groups in the labor force is smaller.

3. Economic Events: Broader economic events, like recessions, can be seen impacting almost all age groups simultaneously, leading to an expanded area during those times.

4. Age Dynamics: The relative contributions of each age group change over time. For instance, the age_55plus_rate layer has grown over the years, possibly reflecting the aging U.S. population and the increasing participation of older individuals in the workforce.

```python
# Calculating the average unemployment rate for the most recent year in the data across age groups
latest_year = df_age['date'].dt.year.max()
avg_unemployment_latest_year = df_age[df_age['date'].dt.year == latest_year][age_columns].mean()

# Plotting the pie chart
plt.figure(figsize=(10, 10))
avg_unemployment_latest_year.plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(age_columns)))
plt.title(f'Average Unemployment Distribution by Age Group for {latest_year}')
plt.ylabel('')  # Removing the default ylabel for clarity
plt.tight_layout()
plt.show()
```

```python
# Plotting the joint distribution plot for men's and women's unemployment rates
sns.jointplot(data=df_sex, x='men_rate', y='women_rate', kind='hex', color='purple', height=8)
plt.suptitle('Joint Distribution of Unemployment Rates: Men vs. Women', y=1.02)
plt.xlabel('Men Unemployment Rate (%)')
plt.ylabel('Women Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```

##### The Joint Distribution plot offers insights into the relationship between men's and women's unemployment rates:

1. Hex Density: The darker hexagons represent areas with a higher density of data points, indicating that for many months, men's and women's unemployment rates were close and aligned.

2. Diagonal Alignment: The data points are clustered along a diagonal line, implying a positive correlation between men's and women's unemployment rates. This means that when unemployment rates for men increased, they generally also increased for women and vice versa.

3. Divergences: While most points are close to the diagonal, some diverge. These represent months where the unemployment rates for men and women differed more substantially.

##### In summary, while men's and women's unemployment rates largely move in tandem, there are periods of divergence, highlighting the unique challenges or opportunities each gender might face in the labor market. Combining these insights with the earlier analyses provides a multifaceted view of the dynamics and nuances of unemployment in the U.S. across different segments of the population.
